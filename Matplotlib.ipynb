# Customizing labels, Ticks, and Grids

```python
import matplotlib.pyplot as plt
import numpy as np

x  = np.array([2007, 2008, 2009, 2010])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20,30])

plt.title("Class size", fontsize=25,
                        family="Arial",
                        fontweight="bold",
                        color="#00eafb")

plt.xlabel("Year", fontsize=20,
                   family="Arial",
                   fontweight="bold",
                   color="#00eafb")

plt.ylabel("Students", fontsize=20,
                   family="Arial",
                   fontweight="bold",
                   color="#00eafb")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x  = np.array([2007, 2008, 2009, 2010])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20,30])

plt.title("Class size", fontsize=20,
                        fontweight="bold",
                        color="#00eafb")

plt.xlabel("Year", fontsize=15,
                   fontweight="bold",
                   color="#00eafb")

plt.ylabel("Students", fontsize=15,
                   fontweight="bold",
                   color="#00eafb")

plt.tick_params(axis="both",
                colors="#009ffb")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xticks(x)

plt.show()
```

# Bar Chart

```python
# Bar Chart = compare categories of data by representing each category with a Bar

import matplotlib.pyplot as plt
import numpy as np

categories = np.array(["Grains", "Fruit", "Vegetable", "Protein", "Dairy", "Sweets"])
values = np.array([4, 3, 2, 5, 3, 1])

plt.bar(categories, values)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = np.array(["Grains", "Fruit", "Vegetable", "Protein", "Dairy", "Sweets"])
values = np.array([4, 3, 2, 5, 3, 1])

plt.bar(categories, values, color="skyblue")

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```

# Pie Chart

```python
# Pie Chart = Circular chart divided into slices to sow percentages of the total.
#             Good for visualizing distribution among categories.

import matplotlib.pyplot as plt
import numpy as np

categories = np.array(["Freshmen", "Sophomores", "Juniors", "Seniors"])
values = np.array([300, 250, 275, 225])

plt.pie(values, labels=categories,
        autopct="%1.1f%%")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = np.array(["Freshmen", "Sophomores", "Juniors", "Seniors"])
values = np.array([300, 250, 275, 225])
colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels=categories,
                autopct="%1.1f%%",
        colors=colors)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = np.array(["Freshmen", "Sophomores", "Juniors", "Seniors"])
values = np.array([300, 250, 275, 225])
colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels=categories,
                autopct="%1.1f%%",
        colors=colors)

plt.title("Anime Colleges")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = np.array(["Freshmen", "Sophomores", "Juniors", "Seniors"])
values = np.array([300, 250, 275, 225])
colors = ["red", "yellow", "blue", "green"]

plt.pie(values, labels=categories,
                autopct="%1.1f%%",       # To show Percentage
                colors=colors,           # To show Colors
                explode=[0, 0, 0, 0.2],  # To Highlight a slice we use explode
                shadow=True,             # To Add a drop Shadow
                startangle=90)           # To rotate you pie chart

plt.title("Anime Colleges")


plt.show()
```

# Scatter Graph

```python
# Scatter graph = Shows the relationship between two variables
#                 Helps to identify a correlation (+, -, None)
#                 Example: Study hours vs. Tset Scores

import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] # Hours studied
y = [55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87] # Grades

plt.scatter(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] # Hours studied
y = [55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87] # Grades

plt.scatter(x, y, color="skyblue",
            alpha = 0.5,     # alpha = for Transparency
            s = 200)         # s = size of dots

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = [0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] # Hours studied
y1 = [55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87] # Grades

x2 = [0, 1, 2, 2, 3, 4, 5, 6, 7, 7, 8] # Hours studied
y2 = [50, 58, 65, 70, 72, 78, 83, 88, 92, 95, 97] # Grades

plt.scatter(x1, y1, color="skyblue",
            alpha = 0.5,     # alpha = for Transparency
            s = 200)         # s = size of dots

plt.scatter(x2, y2, color="red",
            alpha = 0.5,     # alpha = for Transparency
            s = 200)         # s = size of dots

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = [0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8] # Hours studied
y1 = [55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87] # Grades

x2 = [0, 1, 2, 2, 3, 4, 5, 6, 7, 7, 8] # Hours studied
y2 = [50, 58, 65, 70, 72, 78, 83, 88, 92, 95, 97] # Grades

plt.scatter(x1, y1, color="skyblue",
            alpha = 0.5,     # alpha = for Transparency
            s = 200,         # s = size of dots
            label="Class A")

plt.scatter(x2, y2, color="red",
            alpha = 0.5,     # alpha = for Transparency
            s = 200,         # s = size of dots
            label="Class B")

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.legend()   # legend function to show legend

plt.show()
```

# Histograms

```python
# Histogram = A visual representation of the distribution of quantitative data.
#             They group vlues into bins (intervals)
#             and counts how many fall in each range.

import matplotlib.pyplot as plt
import numpy as np

scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10,
         color="lightgreen",
         edgecolor="black")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

scores = np.random.normal(loc=80, scale=10, size=100)
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10,
         color="lightgreen",
         edgecolor="black")

plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("No of Students")

plt.show()
```

# Subplots

```python
# Figure = The entire Canvas

# Ax = A single plot (subplot)

import  matplotlib.pyplot as plt
import numpy as np

figure, axes = plt.subplots(nrows=2, ncols=2)

plt.show()
```

```python
import  matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])

figure, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(x, x*2)

plt.show()
```

```python
import  matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])

figure, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(x, x*2, color="red")

axes[0, 0].set_title("x*2")

plt.show()
```

```python
import  matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])

figure, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(x, x*2, color="red")
axes[0, 0].set_title("x*2")

axes[0, 1].plot(x, x**2, color="blue")
axes[0, 1].set_title("x**2")

plt.show()
```

```python
import  matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])

figure, axes = plt.subplots(nrows=2, ncols=2)

axes[0, 0].plot(x, x*2, color="red")
axes[0, 0].set_title("x*2")

axes[0, 1].plot(x, x**2, color="blue")
axes[0, 1].set_title("x**2")

axes[1, 0].plot(x, x**3, color="green")
axes[1, 0].set_title("x**3")

plt.tight_layout()  # To fit in Shape

plt.show()
```
