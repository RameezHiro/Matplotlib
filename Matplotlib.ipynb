# Intro of Matplotlib

```python
# Matplotlib is a comprehensive library for creating static, animated
 # and interactive visualizations in Python.

import matplotlib.pyplot as plt
import numpy as np

x = np.array([2007, 2008, 2009, 2010])
y = np.array([15, 25, 30, 20])

plt.plot(x, y)

plt.show()
```

# Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2007, 2008, 2009, 2010])
y = np.array([15, 25, 30, 20])

plt.plot(x, y, marker=".")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2007, 2008, 2009, 2010])
y = np.array([15, 25, 30, 20])

plt.plot(x, y, marker=".",
         markersize=30)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2007, 2008, 2009, 2010])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])

line_style = dict(marker=".",
                  markersize=30,
                  markerfacecolor="#f4f803",  # Or mfc (markerfacecolor)
                  markeredgecolor="#f4f803",  # Or mec (markeredgecolor)
                  linestyle="solid",
                  linewidth="4",
                  color="#0078fb")

plt.plot(x, y1, **line_style)
plt.plot(x, y2, **line_style)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2007, 2008, 2009, 2010])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20,30])

line_style = dict(marker=".",
                  markersize=30,
                  markerfacecolor="#f4f803",  # Or mfc (markerfacecolor)
                  markeredgecolor="#f4f803",  # Or mec (markeredgecolor)
                  linestyle="solid",
                  linewidth="4",
                  color="#0078fb")

plt.plot(x, y1, **line_style)
plt.plot(x, y2, **line_style)
plt.plot(x, y3, **line_style)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x  = np.array([2007, 2008, 2009, 2010])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20,30])

line_style = dict(marker=".",
                  markersize=30,
                  markerfacecolor="#f4f803",  # Or mfc (markerfacecolor)
                  markeredgecolor="#f4f803",  # Or mec (markeredgecolor)
                  linestyle="solid",
                  linewidth="4")

plt.plot(x, y1, color="#0078fb", **line_style)
plt.plot(x, y2, color="#52fb00", **line_style)
plt.plot(x, y3, color="#fb4100", **line_style)

plt.show()
```

# Customizing labels, Ticks, and Grids

```python
import matplotlib.pyplot as plt
import numpy as np

x  = np.array([2007, 2008, 2009, 2010])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20,30])

plt.title("Class size", fontsize=25,
                        family="Arial",
                        fontweight="bold",
                        color="#00eafb")

plt.xlabel("Year", fontsize=20,
                   family="Arial",
                   fontweight="bold",
                   color="#00eafb")

plt.ylabel("Students", fontsize=20,
                   family="Arial",
                   fontweight="bold",
                   color="#00eafb")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x  = np.array([2007, 2008, 2009, 2010])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20,30])

plt.title("Class size", fontsize=20,
                        fontweight="bold",
                        color="#00eafb")

plt.xlabel("Year", fontsize=15,
                   fontweight="bold",
                   color="#00eafb")

plt.ylabel("Students", fontsize=15,
                   fontweight="bold",
                   color="#00eafb")

plt.tick_params(axis="both",
                colors="#009ffb")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xticks(x)

plt.show()
```

# Grid Lines

```python
import matplotlib.pyplot as plt
import numpy as np

# Grid() = Helps make plots easier to raed by adding reference lines.

x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.grid()

plt.plot(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.grid(axis="x")

plt.plot(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.grid(axis="y")

plt.plot(x, y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.grid(axis="y",
         linewidth=2,
         color="lightgray",
         linestyle="dashed")

plt.plot(x, y)

plt.show()
```
